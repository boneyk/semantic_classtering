## Определение по должности

```python
prof = pd.read_excel('prof.xlsx',index_col = 'Unnamed: 0')
unique_values = prof[0].str.lower().unique()
unique_values = unique_values.tolist()
unique_values.append('строитель')
prof = unique_values
```

```python
def find_prof(splitted_lemmas):

    for x in prof:

        if x in splitted_lemmas:

            return x
        
df2['по должности-лемме2'] = df2['Леммы'].str.split()
df2['по должности-лемме2'] = df2['по должности-лемме2'].apply(find_prof)
df2['по должности-лемме2'] = df2['по должности-лемме2'].fillna(value= np.nan, inplace=False)
df2[df2['Кластер'] == 49]
```

## Определение по доп. признаку

```python
dop = pd.read_excel('dop_values.xlsx')
keys = dop['синоним'] 
values = dop['фраза']
sm = dict(zip(keys, values))
sm
```

```python
def find_dop(splitted_lemmas):
    return ', '.join([sm[x] for x in splitted_lemmas if x in sm])

df2['по дополнительному признаку2'] = df2['Леммы'].str.split()
df2['по дополнительному признаку2'] = df2['по дополнительному признаку2'].apply(find_dop)
df2['по дополнительному признаку2'] = df2['по дополнительному признаку2'].replace('', np.nan)
```

```python
df2[df2['Кластер'] == 49]
```

## Определение по условиям


```python
condition = {
    'проживание': 'с проживанием',
    'ежедневный': 'с ежедневной оплатой',
}
print(condition)
```

```python
def find_condition(splitted_lemmas):
    return ', '.join([condition[x] for x in splitted_lemmas if x in condition])

df2['по условию2'] = df2['Леммы'].str.split()
df2['по условию2'] = df2['по условию2'].apply(find_condition)
df2['по условию2'] = df2['по условию2'].replace('', np.nan)
```

```python
df2[df2['Кластер'] == 50]
```

## Определение по занятости

```python
busyness = pd.read_excel('busyness.xlsx')
busyness
```

```python
busyness = dict(zip(busyness['фраза'],busyness['синоним']))
busyness
```

```python
def find_busyness(splitted_lemmas):
    return ', '.join([busyness[x] for x in splitted_lemmas if x in busyness])

df2['занятость2'] = df2['Леммы'].str.split()
df2['занятость2'] = df2['занятость2'].apply(find_busyness)
df2['занятость2'] = df2['занятость2'].replace('', np.nan)
```

```python
df2[df2['Кластер'] == 50]
```

```python
# Проверяем, есть ли пустые значения в указанных столбцах
mask = ((df2['по должности-лемме2'].isna())  & \
       (df2['по дополнительному признаку2'].isna()) & \
       (df2['по условию2'].isna()) & \
       (df2['занятость2'].isna()))
df2['общие фразы2'] = np.where(mask,'общие фразы',np.nan)
```
