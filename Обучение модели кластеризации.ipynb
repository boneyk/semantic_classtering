```python
km_clustering = KMeans(
    n_clusters=int(np.round(np.divide(len(df['Леммы']), 5))),
    init='k-means++',
    n_init=10,
    max_iter=300,
    tol=0.0001)
```

## Создание названия кластера

```python
def get_cluster_name(cluster_words):
    # Посчитать частоту каждого слова в кластере
    word_counts = {}
    for words in cluster_words:
        words = words.split()
        for word in words:
            if word in word_counts:
                word_counts[word] += 1
            else:
                word_counts[word] = 1
    
    # Найти 2 самых частых слова и сформировать название кластера
    top_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)[:2]
    cluster_name = ' '.join([word[0] for word in top_words])
    
    return cluster_name

# Применение функции к данным
clasters_name = []
clasters_name = df.groupby('Кластер')['Леммы'].apply(lambda x: get_cluster_name(x.tolist()))
```

```python
for i in range(len(clasters_name)):
    mask = (df['Кластер'] == i)
    df.loc[mask, 'Название_кластера'] = clasters_name[i]
```

```python
df_new = pd.read_csv('dataset_users_query.csv')
df_new = df_new.iloc[3007:3008]
df_new
```

```python
df_new['Фраза_по_словам'] = df_new['query'].str.lower().str.split()
df_new.head()

predlogi = pd.read_excel('spisok_predlogov.xlsx')
predlogi = predlogi['Unnamed: 0'].dropna()
unique_values = predlogi.str.lower().unique()
unique_values = unique_values.tolist()
predlogi = unique_values

def find_predlogi(splitted_phrase):
    for word in splitted_phrase:
        if word in predlogi:
            splitted_phrase.remove(word)
    return splitted_phrase

df_new['Фраза_по_словам'] = df_new['Фраза_по_словам'].apply(find_predlogi)

def find_num(splitted_phrase):
    return [word for word in splitted_phrase if not any(char.isdigit() for char in word)]

df_new['Фраза_по_словам'] = df_new['Фраза_по_словам'].apply(find_num)

def find_word(splitted_phrase):
    return [word for word in splitted_phrase if len(word) > 1 or not word.isalpha()]

df_new['Фраза_по_словам'] = df_new['Фраза_по_словам'].apply(find_word)

df_new['Фраза_по_словам'] = df_new['Фраза_по_словам'].apply(lambda x: ' '.join(x))
df_new.head(100)
```
